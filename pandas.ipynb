<h1>Time series Analysis with pandas</h1>

```python
import pandas
import yfinance

stock = "GS"
start = "2023-09-01"
end = "2023-10-13"

data = yfinance.download(stock, start, end)

dataframe = pandas.DataFrame(data)

dataframe.head(5)
```

<h2>Showcasing pandas methods</h2>

```python
dataframe.head(1)
```

```python
dataframe.info()
```

```python
dataframe.describe()
```

<h1>Boolean Indexing</h1>

```python
dataframe[dataframe["Open"] > 314].head(5)
dataframe[(dataframe["Open"] > 314) & (dataframe["High"] > 315)].head(5)
```

```python
series = dataframe["Close"].head(5)
series
```

<h1>Data manipulation</h1>

```python
dataframe["Open"].iloc[0] = None
dataframe["Close"].iloc[1] = None
dataframe.head(5)
```

```python
cleanDf = dataframe.dropna()
cleanDf.head(5)
```

<h1>Time Series Analysis</h1>

<text>Interpolation - notice closing price for 2023-09-05</text>

```python
dataframe.interpolate(method='linear', inplace=True)
dataframe.head(5)
```

<text>Resampling</text>

```python
weekly = cleanDf.resample('W').mean()
weekly
```

<text>Calculating moving average</text>

```python
dataframe["5MA"] = dataframe["Open"].rolling(window=5).mean()
dataframe
```
